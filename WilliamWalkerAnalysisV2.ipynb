```python
import pandas as pd
```

```python
wwav_df = pd.read_csv("20230222_Total_consolidated.csv")
```

```python
wwav_df.rename(columns={"Unnamed: 0":"Index"},inplace=True)
wwav_df.set_index("Index", inplace=True)
```

```python
wwav_selceted_sku = wwav_df[[
        "SC_variant_id",
        "SC_variant_sku",
        "TVUR_shopify_lineitems_quantity",
        "TVUR_shopify_lineitems_preorder",
        "TVUR_shopify_rolling7days_lineitems_quantity",
        "TVUR_shopify_rolling30days_lineitems_quantity",
        "TVUR_shopify_delta7days_lineitems_quantity",
        "TVUR_shopify_delta30days_lineitems_quantity",
        "TVKR_shopify_rolling7days_lineitems_quantity_lastYear",
        "TVKR_shopify_rolling30days_lineitems_quantity_lastYear",
        "TVUR_shopify_log_lineitems_quantity",
        "TVUR_shopify_rolling7days_lineitems_quantity_trend",
        "TVUR_shopify_rolling7days_lineitems_quantity_season",
        "TVUR_shopify_rolling7days_lineitems_quantity_resid",
        "TVKR_shopify_rolling7days_lineitems_quantity_trend_lastYear",
        "TVKR_shopify_rolling7days_lineitems_quantity_season_lastYear",
        "TVKR_shopify_rolling7days_lineitems_quantity_resid_lastYear",
        "TVKC_year",
        "TVKC_month"
        ]]
```

```python
wwav_selceted_category = wwav_df[[
        "SC_product_category",
        "TVUR_shopify_lineitems_quantity",
        "TVUR_shopify_lineitems_preorder",
        "TVUR_shopify_rolling7days_lineitems_quantity",
        "TVUR_shopify_rolling30days_lineitems_quantity",
        "TVUR_shopify_delta7days_lineitems_quantity",
        "TVUR_shopify_delta30days_lineitems_quantity",
        "TVKR_shopify_rolling7days_lineitems_quantity_lastYear",
        "TVKR_shopify_rolling30days_lineitems_quantity_lastYear",
        "TVUR_shopify_log_lineitems_quantity",
        "TVUR_shopify_rolling7days_lineitems_quantity_trend",
        "TVUR_shopify_rolling7days_lineitems_quantity_season",
        "TVUR_shopify_rolling7days_lineitems_quantity_resid",
        "TVKR_shopify_rolling7days_lineitems_quantity_trend_lastYear",
        "TVKR_shopify_rolling7days_lineitems_quantity_season_lastYear",
        "TVKR_shopify_rolling7days_lineitems_quantity_resid_lastYear",
        "TVKC_year",
        "TVKC_month"
        ]]
```

```python
wwav_selceted_product_id = wwav_df[[
        "SC_product_id",
        "TVUR_shopify_lineitems_quantity",
        "TVUR_shopify_lineitems_preorder",
        "TVUR_shopify_rolling7days_lineitems_quantity",
        "TVUR_shopify_rolling30days_lineitems_quantity",
        "TVUR_shopify_delta7days_lineitems_quantity",
        "TVUR_shopify_delta30days_lineitems_quantity",
        "TVKR_shopify_rolling7days_lineitems_quantity_lastYear",
        "TVKR_shopify_rolling30days_lineitems_quantity_lastYear",
        "TVUR_shopify_log_lineitems_quantity",
        "TVUR_shopify_rolling7days_lineitems_quantity_trend",
        "TVUR_shopify_rolling7days_lineitems_quantity_season",
        "TVUR_shopify_rolling7days_lineitems_quantity_resid",
        "TVKR_shopify_rolling7days_lineitems_quantity_trend_lastYear",
        "TVKR_shopify_rolling7days_lineitems_quantity_season_lastYear",
        "TVKR_shopify_rolling7days_lineitems_quantity_resid_lastYear",
        "TVKC_year",
        "TVKC_month"
        ]]
```

```python
sku = wwav_selceted_sku.groupby(by=["SC_variant_sku", "SC_variant_id", "TVKC_year", "TVKC_month"], as_index=False).sum()
```

```python
category = wwav_selceted_category.groupby(by=["SC_product_category", "TVKC_year", "TVKC_month"], as_index=False).sum()
```

```python
product_id = wwav_selceted_product_id.groupby(by=["SC_product_id","TVKC_year", "TVKC_month"], as_index=False).sum()
```

```python
writer = pd.ExcelWriter('WilliamWalkerAnalysis.xlsx', engine='xlsxwriter')
```

```python
sku.to_excel(writer,sheet_name="sku_level")
category.to_excel(writer,sheet_name="category_level")
product_id.to_excel(writer,sheet_name="product_id_level")

writer.save()
```
